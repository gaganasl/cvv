```python
conversation_data = {

    "hello": "Hi there! How can i help you today?",
    "hi":"Hello! How are you?",
    "how are you":"I'm just a bot, but I'm doing great! How about you?",
    "What is computer vision":"Computre vision is a field of AI that allows computer to see",
     "What is ai": "AI stands for Artificial Intellegience, Which enables machines to perform ",
    "What is deep learning":"Deep Learning is a subject of ai that uses neural network",
    "bye":"Goodbye! Have a nice day!",
    "thank you":"you're welcome! happy to help"
}
```

```python
def sample_chat(user_input, chat_history=[]):
  user_input_lower = user_input.lower()
  response = None # Corrected 'none' to 'None'

  for key in conversation_data:
    if key in user_input_lower:
      response = conversation_data[key]
      break # Found a match, exit loop

  if response is None: # If no match was found after checking all keys
    response = "I'm sorry, I don't understand that. Can you ask something else?"

  chat_history.append(("you", user_input))
  chat_history.append(("bot", response))

  return response, chat_history
```

```python
print("welcome to Mini Conversation AI")
print("Type 'exit', or 'quit',to end the chat.\n")

chat_history = []

while True:
  user_input = input("you: ")

  if user_input.lower() in ["exit","quit"]:
    print("chat ended.Goodbye!")
    break

  # These lines are now correctly unindented to be part of the main loop
  bot_response, chat_history = sample_chat(user_input, chat_history)
  print("Bot:", bot_response)
```

```python
print("\n======full  chat history ======")
for speaker, message in chat_history:
  print(f"{speaker}: {message}")

# Useful for debugging or showing what happened in the session
```
